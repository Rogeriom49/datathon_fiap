```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Rogeriom49/datathon_fiap/refs/heads/main/datasets/alunos_2.csv')
```

```python
# Criação dos dataframes auxiliares
df_pedras = df.groupby('ANO')['PEDRA'].value_counts().unstack().fillna(0)
means_by_year = round(df.groupby('ANO')[['IAA', 'IEG', 'IPS', 'IDA', 'IPP', 'IPV', 'IAN']].mean().T,2)
```

```python
df_pedras
```

```python
means_by_year
```

```python
# Análise da quantidade de pédras por anos
df_pedras.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Total de Pedras por Ano')
plt.xlabel('Ano')
plt.ylabel('Total de Pedras')

for i in range(len(df_pedras)):
    plt.text(i, 0, df_pedras.iloc[i].to_string(), ha='center', va='bottom')

```

```python
# Análise dá média dos indices por ano
means_by_year.plot(kind='bar', figsize=(12, 6))
plt.title('Médias de indices por Ano')
plt.xlabel('Ano')
plt.ylabel('Média')
```
